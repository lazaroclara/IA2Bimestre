```python
# Biblioteca para manipulação de tabelas
import pandas as pd

# Biblioteca para gráficos
import matplotlib.pyplot as plt

# Base Wine Dataset do scikit-learn
from sklearn.datasets import load_wine

# Algoritmo Apriori e geração das regras de associação
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Carrega a base Wine
wine = load_wine()

# Cria um DataFrame com os atributos químicos dos vinhos
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# Adiciona a classe do vinho como uma coluna categórica
df['classe_vinho'] = [wine.target_names[i] for i in wine.target]

# Exibe as primeiras linhas da base
df.head()
```

```python
# Informações gerais da base
print('Quantidade de amostras:', df.shape[0])
print('Quantidade de atributos:', len(wine.feature_names))
print('Classes:', wine.target_names)
```

```python
# Copia a base original para não alterar o DataFrame inicial
df_categorico = pd.DataFrame()

# Para cada atributo numérico, divide os valores em 3 faixas: baixo, médio e alto
for coluna in wine.feature_names:
    df_categorico[coluna] = pd.qcut(
        df[coluna],
        q=3,
        labels=['baixo', 'medio', 'alto'],
        duplicates='drop'
    )

# Mantém a classe do vinho como uma variável categórica
df_categorico['classe_vinho'] = df['classe_vinho']

# Exibe os dados categorizados
df_categorico.head()
```

```python
# Converte as categorias em matriz binária usando One-Hot Encoding
# Cada coluna passa a representar um item possível
transacoes = pd.get_dummies(df_categorico)

# Converte os valores para booleanos, formato recomendado pela mlxtend
transacoes = transacoes.astype(bool)

# Exibe as primeiras transações
transacoes.head()
```

```python
# Mostra o tamanho da matriz transacional
print('Quantidade de transações:', transacoes.shape[0])
print('Quantidade de itens gerados:', transacoes.shape[1])
```

```python
# Define o suporte mínimo
suporte_minimo = 0.20

# Aplica o algoritmo Apriori para encontrar itemsets frequentes
itemsets_frequentes = apriori(
    transacoes,
    min_support=suporte_minimo,
    use_colnames=True
)

# Adiciona uma coluna com o tamanho do itemset
itemsets_frequentes['tamanho'] = itemsets_frequentes['itemsets'].apply(lambda x: len(x))

# Ordena do maior suporte para o menor
itemsets_frequentes = itemsets_frequentes.sort_values(by='support', ascending=False)

# Exibe os itemsets frequentes
itemsets_frequentes.head(20)
```

```python
# Seleciona os 10 itemsets mais frequentes
top_itemsets = itemsets_frequentes.head(10).copy()

# Converte os itemsets para texto para facilitar a visualização
top_itemsets['itemsets_texto'] = top_itemsets['itemsets'].apply(lambda x: ', '.join(list(x)))

plt.figure(figsize=(10, 6))
plt.barh(top_itemsets['itemsets_texto'], top_itemsets['support'])
plt.xlabel('Suporte')
plt.ylabel('Itemsets')
plt.title('Top 10 Itemsets Frequentes - Wine Dataset')
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```

```python
# Define a confiança mínima
confianca_minima = 0.80

# Gera as regras de associação
regras = association_rules(
    itemsets_frequentes,
    metric='confidence',
    min_threshold=confianca_minima
)

# Ordena as regras pela confiança e pelo lift
regras = regras.sort_values(by=['confidence', 'lift'], ascending=False)

# Exibe as principais colunas
regras[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(20)
```

```python
# Função para transformar frozenset em texto
def formatar_itemset(itemset):
    return ', '.join(list(itemset))

# Cria uma tabela mais fácil de ler
regras_formatadas = regras.copy()
regras_formatadas['SE'] = regras_formatadas['antecedents'].apply(formatar_itemset)
regras_formatadas['ENTAO'] = regras_formatadas['consequents'].apply(formatar_itemset)

# Seleciona as colunas principais
regras_formatadas = regras_formatadas[[
    'SE',
    'ENTAO',
    'support',
    'confidence',
    'lift'
]]

# Exibe as 20 primeiras regras formatadas
regras_formatadas.head(20)
```

```python
# Filtra regras em que o consequente contém a classe do vinho
regras_classe = regras[
    regras['consequents'].apply(lambda x: any('classe_vinho' in item for item in x))
].copy()

# Formata as regras de classe
regras_classe['SE'] = regras_classe['antecedents'].apply(formatar_itemset)
regras_classe['ENTAO'] = regras_classe['consequents'].apply(formatar_itemset)

# Exibe as principais regras associadas às classes
regras_classe[[
    'SE',
    'ENTAO',
    'support',
    'confidence',
    'lift'
]].head(20)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(regras['support'], regras['confidence'])
plt.xlabel('Suporte')
plt.ylabel('Confiança')
plt.title('Regras de Associação - Suporte x Confiança')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(regras['confidence'], regras['lift'])
plt.xlabel('Confiança')
plt.ylabel('Lift')
plt.title('Regras de Associação - Confiança x Lift')
plt.grid(True)
plt.show()
```

```python
# Teste com outros parâmetros
novo_suporte = 0.15
nova_confianca = 0.75

itemsets_teste = apriori(
    transacoes,
    min_support=novo_suporte,
    use_colnames=True
)

regras_teste = association_rules(
    itemsets_teste,
    metric='confidence',
    min_threshold=nova_confianca
)

print('Quantidade de itemsets frequentes:', len(itemsets_teste))
print('Quantidade de regras geradas:', len(regras_teste))

regras_teste[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```
