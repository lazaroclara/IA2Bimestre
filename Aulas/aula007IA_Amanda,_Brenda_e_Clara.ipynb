```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report

from tensorflow.keras.datasets import fashion_mnist
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

print("Formato original das imagens:", X_train.shape)
```

```python
plt.imshow(X_train[0], cmap='gray')
plt.title(f"Classe: {y_train[0]}")
plt.show()
```

```python
X_train = X_train.reshape(len(X_train), -1)
X_test = X_test.reshape(len(X_test), -1)

print("Formato após transformação:", X_train.shape)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
from sklearn import svm

modelo = svm.SVC(kernel='linear')

print("Treinando o modelo...")
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
print("Acurácia:", accuracy_score(y_test, y_pred))

print("\nRelatório de classificação:")
print(classification_report(y_test, y_pred))
```

```python
indice = 10

plt.imshow(X_test[indice].reshape(28,28), cmap='gray')
plt.title(f"Real: {y_test[indice]} | Previsto: {y_pred[indice]}")
plt.show()
```
